```python
def random_centroids(data,k):
  centroids =[]
  for i in range (k):
   centroid =data.apply(lambda x: float(x.sample()))
   centroids.append(centroid)
  return pd.concat(centroids,axis=1)
```

```python
centroids = random_centroids(data,3)
centroids
```

```python
def get_labels(data,centroids):
  distances=centroids.apply(lambda x: np.sqrt(((data-x)**2).sum(axis=1)))
  return distances.idxmin(axis=1)
```

```python
labels=get_labels(data,centroids)
labels.value_counts()    # Assigning dinos to a cluster based on random centroids
```

```python
def new_centroids(data,labels,k):
  return data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T     # Calculating gEometric mean from the centroid
```

```python
from sklearn.decomposition import PCA
from IPython.display import clear_output
```

```python
def calculate_inertia(data, centroids, labels):
    # Calculate inertia as the sum of squared distances of points to their cluster centroids.
    inertia = 0
    for label in centroids.columns:
        cluster_points = data[labels == label]
        centroid = centroids[label]
        inertia += ((cluster_points - centroid) ** 2).sum().sum()
    return inertia

def elbow_plot(data, k_range, max_iterations=100):
    #Generate an elbow plot to find the optimal value of k
    inertias = []

    for k in k_range:
        centroids = random_centroids(data, k)
        old_centroids = pd.DataFrame()
        iteration = 0

        # Run k-means for this value of k
        while iteration < max_iterations and not centroids.equals(old_centroids):
            old_centroids = centroids
            labels = get_labels(data, centroids)
            centroids = new_centroids(data, labels, k)
            iteration += 1

        # Calculate inertia for this k
        labels = get_labels(data, centroids)
        inertia = calculate_inertia(data, centroids, labels)
        inertias.append(inertia)

    # Plot the elbow curve
    plt.figure(figsize=(8, 5))
    plt.plot(k_range, inertias, marker='o', linestyle=':')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia (Sum of Squared Distances)')
    plt.title('Elbow Method for Optimal k')
    plt.xticks(ticks=list(k_range))
    plt.show()

# Example usage
k_range = range(1, 11)  # Test k values from 1 to 10
elbow_plot(data, k_range)
```

From the Elbow curve we can clearly see that the elbow point is 3

```python
def plot_clusters(data,labels,centroids,iteration):
  pca = PCA(n_components=2)
  data_2d = pca.fit_transform(data)
  centroids_2d=pca.transform(centroids.T)
  clear_output(wait=True)
  plt.title(f'Iteration {iteration}')
  plt.scatter(x=data_2d[:,0],y=data_2d[:,1],c=labels)
  plt.scatter(x=centroids_2d[:,0],y=centroids_2d[:,1])
  plt.show()
```

```python
max_iterations=100
k=3
centroids=random_centroids(data,k)
old_centroids=pd.DataFrame()
iteration=1
while iteration<max_iterations and not centroids.equals(old_centroids):
  old_centroids=centroids
  labels=get_labels(data,centroids)
  centroids=new_centroids(data,labels,k)
  plot_clusters(data,labels,centroids,iteration)
  iteration+=1
```
