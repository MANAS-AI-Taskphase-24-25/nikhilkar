```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Threats (1).csv')
```

```python
data.head()
```

```python
data['service'] = data['service'].replace('-', 'dns')   #Frequency encoding the missing service rows
```

```python
from sklearn.preprocessing import LabelEncoder
le_proto = LabelEncoder()
le_service = LabelEncoder()
le_state = LabelEncoder()
le_attack_cat = LabelEncoder()
```

```python
data['proto'] = le_proto.fit_transform(data['proto'])
data['service'] = le_service.fit_transform(data['service'])
data['state'] = le_state.fit_transform(data['state'])
data['attack_cat'] = le_attack_cat.fit_transform(data['attack_cat'])
data.head()
```

```python
X= data.drop(['attack_cat'],axis=1)    # Separating output label from feature dataset
y= data['attack_cat']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=44,stratify=y)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(max_depth= None,max_features=None,min_samples_leaf=2,min_impurity_decrease=0.000001,ccp_alpha=0.0001,random_state=44,splitter='best')
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
 # Probability of each class
pp=model.predict_proba(X_test)
print(pp)
```

```python
predictions=model.predict(X_test)
print(predictions)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, predictions,average='weighted')
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, predictions,average='weighted')
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, predictions,average='weighted')
```
