```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Threats (1).csv')
```

```python
data.head()
```

```python
data=data.drop(['Unnamed: 0'],axis=1)
```

```python
titles = list(data)[1:-1]
```

```python
data['service'] = data['service'].replace('-', 'dns')
```

```python
from sklearn.preprocessing import LabelEncoder
le_proto = LabelEncoder()
le_service = LabelEncoder()
le_state = LabelEncoder()
le_attack_cat = LabelEncoder()
```

```python
data['proto'] = le_proto.fit_transform(data['proto'])
data['service'] = le_proto.fit_transform(data['service'])
data['state'] = le_proto.fit_transform(data['state'])
data['attack_cat'] = le_proto.fit_transform(data['attack_cat'])
data.head()
```

```python
X= data.drop(['attack_cat'],axis=1)
y= data['attack_cat']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=40)
```

```python

from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators =500, random_state = 42,max_depth=12,min_samples_split=5,max_features='log2')
rf_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = rf_model.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_pred, average='weighted')
```

```python
from sklearn.metrics import matthews_corrcoef
matthews_corrcoef(y_test, y_pred)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, y_pred, average='weighted')
```
